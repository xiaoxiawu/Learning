```python
import io
import re
import matplotlib.pyplot as plt
get_ipython().magic(u'matplotlib inline')
import numpy as np

#filename='translate.txt'
def read_output(filename):
    lines=[line.rstrip('\n') for line in io.open(filename)]

    iter_num=[]
    perple=[]
    perple_bucket0=[]
    perple_bucket1=[]
    perple_bucket2=[]
    perple_bucket3=[]
    for i in range(len(lines)-2):
        element=lines[i]
    #print element
    #global step 1600 learning rate 0.1000 step-time 0.49 perplexity 155.53
        pattern = 'global step \d+ learning rate .\d*.\d+ step-time .\d*.\d+ perplexity .\d*.\d+'
        z=re.findall(pattern,element)
        if len(z)!=0:
            num=z[0].split(',')[0].split(' ')##interation number
            iter_num.append(np.float(num[2]))
            perple.append(np.float(num[len(num)-1]))
            element=lines[i+1]
        #print element
            z=re.findall(' bucket \d+ perplexity .\d*.\d+',element)
            num=z[0].split(',')[0].split(' ')
            perple_bucket0.append(np.float(num[len(num)-1]))
            element=lines[i+2]
        #print element
            z=re.findall(' bucket \d+ perplexity .\d*.\d+',element)
            num=z[0].split(',')[0].split(' ')
            perple_bucket1.append(np.float(num[len(num)-1]))
            element=lines[i+3]
        #print element
            z=re.findall(' bucket \d+ perplexity .\d*.\d+',element)
            num=z[0].split(',')[0].split(' ')
            perple_bucket2.append(np.float(num[len(num)-1]))
            element=lines[i+3]
        #print element
            z=re.findall(' bucket \d+ perplexity .\d*.\d+',element)
            num=z[0].split(',')[0].split(' ')
            perple_bucket3.append(np.float(num[len(num)-1]))
    return  iter_num,perple,perple_bucket0,perple_bucket1,perple_bucket2,perple_bucket3
filename='translate.txt'
x=read_output(filename)
filename1='nopretrainingtranslate.txt'
x1=read_output(filename1)
        #z=re.findall('(\w+)\s*=\s*(\d*)',element)
plt.plot(x[0],x[1],c='r',label=filename)
plt.plot(x1[0],x1[1],c='b',label=filename1)
plt.legend()
plt.show()
plt.plot(x[0],x[2],c='r',label=filename)
plt.plot(x1[0],x1[2],c='b',label=filename1)
plt.legend()
plt.show()
plt.plot(x[0],x[3],c='r',label=filename)
plt.plot(x1[0],x1[3],c='b',label=filename1)
plt.legend()
plt.show()
plt.plot(x[0],x[4],c='r',label=filename)
plt.plot(x1[0],x1[4],c='b',label=filename1)
plt.legend()
plt.show()
```
